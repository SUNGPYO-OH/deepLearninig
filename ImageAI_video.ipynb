```
from imageai.Detection import VideoObjectDetection
import os

detector = VideoObjectDetection()

path = "/content/drive/My Drive/Colab Notebooks/imageai_video/"

detector.setModelTypeAsYOLOv3()
detector.setModelPath(path + "yolo.h5")
detector.loadModel()

video_path = detector.detectObjectsFromVideo(input_file_path=path + "movie.mp4",
                                             output_file_path=path + "detected_video2",
                                             frames_per_second=20, log_progress=True)
print(video_path)
# It takes long long time
```
